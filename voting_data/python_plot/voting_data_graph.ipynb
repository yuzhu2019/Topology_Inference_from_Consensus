```python
import numpy as np
import pandas as pd
import scipy.io as scio
import matplotlib.pyplot as plt
import networkx as nx

data1 = scio.loadmat('A100.mat') # top 100 edges 
data2 = scio.loadmat('A130.mat') # top 130 edges 
data3 = scio.loadmat('A150.mat') # top 150 edges 
data4 = scio.loadmat('A200.mat') # top 200 edges 
data5 = scio.loadmat('A250.mat') # top 250 edges 
data6 = scio.loadmat('A300.mat') # top 300 edges 
```

```python
G1 = nx.from_numpy_matrix(data1['A'])
G2 = nx.from_numpy_matrix(data2['A'])
G3 = nx.from_numpy_matrix(data3['A'])
G4 = nx.from_numpy_matrix(data4['A'])
G5 = nx.from_numpy_matrix(data5['A'])
G6 = nx.from_numpy_matrix(data6['A'])
```

```python
#type1 = [1,3,5,6,7,8,9,13,18,23,34,43,46,47,48,50] # 16 blue
#type2 = [2,4,10,11,12,14,15,19,21,26,37,39,41,42]  # 14 mixed
#type3 = [16,17,20,22,24,25,27,28,29,30,31,32,33,35,36,38,40,44,45,49] # 20 red
type1 = [1,3,5,7,8,9,13,18,23,34,43,46,47,48,50] # 15 blue
type2 = [4,10,11,12,14,15,19,21,26,37,39,41,42]  # 13 mixed
type3 = [16,17,20,22,24,25,27,28,29,30,31,32,33,35,36,38,40,44,45,49] # 20 red
type4 = [6] # blue & independent
type5 = [2] # red & independent

node_color = [0 for i in range(50)]
for i in np.arange(15):
    node_color[type1[i]-1] = 'skyblue'
for i in np.arange(13):
    node_color[type2[i]-1] = 'yellow'
for i in np.arange(20):
    node_color[type3[i]-1] = 'r'
for i in np.arange(1):
    node_color[type4[i]-1] = 'cyan'
for i in np.arange(1):
    node_color[type5[i]-1] = 'orange'
```

```python
state_name = ["CT","ME","MA","NH","RI","VT","DE","NJ","NY","PA","IL","IN","MI","OH","WI","IA","KS","MN","MO","NE","ND","SD","VA","AL","AR","FL","GA","LA","MS","NC","SC","TX","KY","MD","OK","TN","WV","AZ","CO","ID","MT","NV","NM","UT","WY","CA","OR","WA","AK","HI"]
labels={}
for i in np.arange(50):
    labels[i] = state_name[i]
```
